```python
import pandas as pd

validation = pd.read_csv("data/new_validation.csv")

validation = validation.sort_values(["datadate","tic"])
validation.index = validation.datadate.factorize()[0]

from env.EnvMultipleStock_validation import StockEnvValidation

env_val = StockEnvValidation(validation)

from stable_baselines import A2C

model_a2c = A2C.load("trained_models/2026-03-12_11-28-26/A2C.zip")

obs = env_val.reset()

for i in range(len(validation.index.unique())):

    action, _ = model_a2c.predict(obs)
    obs, rewards, dones, info = env_val.step(action)

    if dones:
        break
        
account_value = env_val.asset_memory

df_val = pd.DataFrame(account_value)
df_val.columns = ["account_value"]

df_val.to_csv("results/a2c_validation_value.csv", index=False)

returns = df_val.account_value.pct_change()

sharpe = (returns.mean()/returns.std()) * (252**0.5)

print(sharpe)
```

```python
import pandas as pd

validation = pd.read_csv("data/new_validation.csv")

validation = validation.sort_values(["datadate","tic"])
validation.index = validation.datadate.factorize()[0]

from env.EnvMultipleStock_validation import StockEnvValidation

env_val = StockEnvValidation(validation)

from stable_baselines import DDPG

model_ppdg = DDPG.load("trained_models/2026-03-16_15-20-13/DDPG.zip")

obs = env_val.reset()

for i in range(len(validation.index.unique())):

    action, _ = model_ppdg.predict(obs)
    obs, rewards, dones, info = env_val.step(action)

    if dones:
        break
        
account_value = env_val.asset_memory

df_val = pd.DataFrame(account_value)
df_val.columns = ["account_value"]

df_val.to_csv("results/DDPG_validation_value.csv", index=False)

returns = df_val.account_value.pct_change()

sharpe = (returns.mean()/returns.std()) * (252**0.5)

print(sharpe)
```

```python
import pandas as pd
import numpy as np

a2c_val = pd.read_csv("results/a2c_validation_value.csv")
ppo_val = pd.read_csv("results/ppo_validation_value.csv")
ddpg_val = pd.read_csv("results/ddpg_validation_value.csv")

def sharpe_ratio(df):
    returns = df.account_value.pct_change().dropna()
    return np.sqrt(252) * returns.mean() / returns.std()

print("A2C Sharpe:", sharpe_ratio(a2c_val))
print("PPO Sharpe:", sharpe_ratio(ppo_val))
print("DDPG Sharpe:", sharpe_ratio(ddpg_val))
```
