```python
import pandas as pd
import numpy as np

a2c_val = pd.read_csv("results/a2c_validation_value.csv")
ppo_val = pd.read_csv("results/ppo_validation_value.csv")
ddpg_val = pd.read_csv("results/ddpg_validation_value.csv")

def sharpe_ratio(df):
    returns = df.account_value.pct_change().dropna()
    return np.sqrt(252) * returns.mean() / returns.std()

print("A2C Sharpe:", sharpe_ratio(a2c_val))
print("PPO Sharpe:", sharpe_ratio(ppo_val))
print("DDPG Sharpe:", sharpe_ratio(ddpg_val))
```
